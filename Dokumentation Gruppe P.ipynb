# 1. Abstract

In dieser Arbeit geht es um den aufbauenden Prozess, ein Hotelreservierungssystem für Schweizer Hotels zu erstellen. Der Fokus der Arbeit liegt auf der Datenbankstruktur und wie man mittels SQL-Statement gewünschte Informationen über das System erhält. Ziel der Arbeit ist es eine Datenbank für künftige Hotelreservierungen zu schaffen, welche alle relevanten Daten von Hotels, Zimmern, Kunden, Buchungen, Rechnungen und Zahlungen erfasst und auch erweitert werden kann. Die Organisation des Projekts wurde in vier Sprints unterteilt. Dabei waren Werkzeuge, wie Sprint Weekly, Reviews, Sprint Retrospektiven und das Kanban-Board, um die Aufgabe zu definieren und zuzuteilen sehr hilfreich. Die Sprint Weekly Reviews und Retrospektiven wurden im One Note dokumentiert.  Die Projektdaten wurden mittels GitHub synchronisiert, so konnte man stets eine Versionsgültigkeit garantieren. 

Im ersten Sprint geht es darum, mittels Excel Tabelle die ersten Daten zu erfassen und grob zu gliedern. Im zweiten Sprint werden im Visual Paradigm Tabellen und Attribute Anhand der Excel Tabellen mit ihren jeweiligen Abhängigkeiten zueinander definiert. Im dritten Sprint wird die SQLite Datenbank erstellt, mit Daten befüllt und anschliessend mittels SQL-Statements diverse User Stories abgebildet. Der vierte Sprint, die Arbeit ins Deepnote zu integrieren wurde im Verlauf des zweiten Sprints gestartet und nach und nach ergänzt. Wir konzentrieren uns darum im vierten Sprint auf die Finalisierung der Arbeit. Die vorgegebenen User Stories dienen als Überprüfung, ob die Datenbank sinnvoll angewendet werden kann und konsistent ist. Sowohl als Administrator der Datenbank, als auch von Kunden, die gerne ein Aufenthalt im Hotel buchen wollen.

# 3. Aufgabe

In einem renommierten Schweizer Skiort wird ein neues Hotelreservierungssystem eingeführt. 
Du (bzw. ihr als Gruppe) übernimmst die Rolle des Business Analysten. Deine Hauptaufgabe 
besteht darin, die Anforderungen der Gäste (auch als Benutzer bezeichnet) an dieses System 
zu erfassen und die relevanten Daten zu bestimmen, die für die Erfüllung dieser 
Anforderungen notwendig sind. Der Schwerpunkt liegt dabei ausschliesslich auf den zu 
verarbeitenden Daten, nicht auf funktionalen Anforderungen wie Login-Prozesse oder Aspekte 
der Benutzererfahrung (UX-User Experience). Das Ziel ist es, ein standardisiertes System zu 
entwickeln, das für alle Hotelbuchungen in Schweizer Hotels eingesetzt werden kann.

Ein Besitzer kann ein oder mehrere Hotels in der Schweiz besitzen. Jedes Hotel offeriert 
mindestens ein Zimmer für Kunden. Jedes Zimmer hat einen bestimmten Zimmerpreis und 
einen Typ (z.B. EZ, DZ, Studio, …). Für ein Hotel können beliebig viele Buchungen durch 
einen Kunden vorgenommen werden. Bei einer Buchung wird das Startdatum, das Enddatum
des Aufenthaltes sowie das Buchungsdatum im System vermerkt. Eine Buchung kann auch 
wieder storniert werden. Für jede realisierte Buchung wird nach Beendigung des Aufenthaltes 
eine Rechnung fällig (Ausnahme ist eine stornierte Buchung). Die Rechnung besteht i.d.R. aus 
mehreren Positionen wie Nächtigung, Getränke, Reinigung, Skipass etc. Für eine Rechnung 
werden auch die Zahlungsart wie Kreditkarte (VISA, AMEX, …), Twint, Paypal, 
Einzahlungsschein, Cash sowie das Rechnungsdatum und die Währung vermerkt. Es kann 
auch vorkommen, dass eine Rechnung nie bezahlt wird. Hotels, sowie deren Kunden haben 
jeweils genau eine Adresse mit Strasse (inkl. Nummer), Postleitzahl, Ortsname und Land.

## 5.1 Scrum

### 5.1.1 Was ist Scrum?

Scrum ist eine agile Projektmanagementmethode, welche häufig in der Softwareentwicklung eingesetzt wird. Es basiert auf iterativer Arbeit, um komplexe Projekte effizient und flexibel zu steuern. Das Ziel ist mittels guter Kommunikation Projektabschnitte in Form von Sprints erfolgreich abzuschliessen und damit ein Teilprodukt zu realisieren.

### 5.1.2 Grundprinzipien von Scrum

- Transparenz: Alle Aspekte des Prozesses müssen für alle Beteiligten sichtbar sein. Dies wird mittels einem für alle zugänglichem Kanban Board im Teams umgesetzt. 

- Überprüfung: Regelmässige Inspektionen des Fortschritts und der Prozesse ermöglichen Anpassungen. In Form von Dailys oder Weeklys, welche festgehalten werden, können solche Überprüfungen umgesetzt werden. 

- Anpassung: Basierend auf den Überprüfungen werden Prozesse und Arbeitsweisen kontinuierlich optimiert. 

### 5.1.3 Rollen in Scrum

- Product Owner: Der Product Owner, also die Dozenten überprüfen die Arbeiten eines Sprints und nehmen sie ab. Während des Sprints unterstützen sie, um auf eine optimale und zur Aufgabe passenden Lösung zu kommen.

- Scrum Master: Unterstützt das Team dabei, Scrum richtig anzuwenden und ist verantwortlich für Sprints, deren Tickets sowie die Umsetzung von diesen.

- Entwicklungsteam: Selbstorganisiertes Team, das die Arbeit in Sprints erledigt und ein fertiges Produkt liefert. Dies wäre in unserem Fall die ganze Gruppe.

### 5.1.4 Ereignisse in Scrum

- Sprint: Ein Zeitrahmen von normalerweise ein bis vier Wochen, in dem ein fertiges Produktinkrement erstellt wird. So ist es auch bei unseren Sprints der Fall, denn der Sprint eins und vier hatte zwei Wochen, während der Sprint zwei und drei jeweils vier Wochen lang waren.

- Weekly: Bei Scrum gibt es Dailys oder Weeklys, bei denen sich die Gruppe über die Fortschritte austauscht. In unserem Fall machen Weeklys mehr Sinn, da wir uns wöchentlich sehen. 

- Sprint Review: Am Ende des Sprints präsentiert das Team das Ergebnis den jeweiligen Product Owners und sammelt Feedback.

- Sprint Retrospektive: Ist zur Reflexion über den vergangenen Sprint da, um Prozesse zu verbessern.

### 5.1.5 Artefakte in Scrum

- Product Backlog: Eine priorisierte Liste aller gewünschten Produktmerkmale und Funktionen, was in unserem Fall die Tabelle ist, welche wir in grober Form als Semesterplan erhalten haben. Dabei können die Übungen, was getan wird und was schlussendlich geliefert werden muss nachgelesen werden. Hier ein Beispiel anhand des ersten Sprints:

### 5.1.6 Kanban Board

Ein Kanban-Board ist ein visuelles Werkzeug zur Verwaltung von Arbeitsprozessen und zur Steigerung der Effizienz in Projekten. Es ermöglicht Teams, Aufgaben und ihren Fortschritt klar und übersichtlich darzustellen.

In unserem Projekt haben wir das Kanban-Board in die folgenden Listen unterteilt:

- To-Do: Hier sind alle anstehenden Aufgaben gesammelt, die noch nicht begonnen wurden.

- Doing: Aufgaben, an denen wir aktuell arbeiten, werden in diese Liste verschoben. Dabei haben wir auch beachtet, dass die Liste absteigend aufgebaut ist, um ein Schritt für Schritt vorgehen zu ermöglichen.

- Done: Abgeschlossene Aufgaben landen hier. Sobald ein Sprint beendet ist, setzen wir die Tickets in dieser Liste auf "Erledigt", wodurch sie nicht mehr angezeigt werden, dies ermöglicht ein übersichtliches Board. Das sieht wie folgt aus:

![Kanban_Done](/bilder/Kanban_done.png)

Wir haben die Aufgaben bewusst nicht einzelnen Teammitgliedern zugeordnet, da wir die meisten Arbeiten gemeinsam erledigt haben. Dies förderte die Zusammenarbeit und stellt sicher, dass jeder im Team über alle Aspekte des Projekts informiert ist.

Um den Fortschritt innerhalb der einzelnen Sprints nachverfolgen zu können, erstellen wir ein Label pro Sprint. Diese Labels dienen der Trennung und Übersicht, sodass man weiss, welche Aufgabe zu welchem Sprint gehört. So haben wir für für die Sprints der Reihe nach von Sprint eins bis vier: blaue, pinke, gelbe und grüne Labels verwendet.

## 5.2 GitHub

Git ist ein weit verbreitetes Versionsverwaltungssystem, das es ermöglicht, Änderungen an Dateien und Quellcode nachzuverfolgen. GitHub baut auf Git auf und bietet eine Online-Plattform zur Speicherung, Freigabe und gemeinsamen Bearbeitung von Projekten.

Wir haben uns entschieden, Git und GitHub für unser Projekt zu nutzen, um unsere Projektdaten sicher zu speichern und zu versionieren. Um die Bedienung noch einfacher zu gestalten, setzen wir GitHub Desktop ein. Diese benutzerfreundliche Anwendung ermöglicht es uns, GitHub direkt von einer Desktop App aus zu nutzen, ohne tiefgehende Kenntnisse der Kommandozeile zu benötigen. GitHub Desktop ist leicht zu bedienen und erleichtert uns den Umgang mit Versionen sowie die Synchronisation unserer Dateien mit dem Online-Repository.

Dies ermöglicht uns, jederzeit auf frühere Versionen unserer Dateien zuzugreifen und die Entwicklungshistorie transparent nachzuvollziehen. Im Vergleich zu anderen Tools wie Microsoft Teams, bietet GitHub spezialisierte Funktionen für die Versionskontrolle, was den Umgang mit Änderungen und die Übersicht deutlich erleichtert. Daher haben wir bewusst auf Teams verzichtet, da es mit GitHub und insbesondere mit GitHub Desktop einfacher geht. Die Entität A steht mit einer oder keiner Beziehung zu Entität B. Entität B steht mit nur einer Beziehung zu Entität A.

![Zero or Many](/bilder/Zero_or_Many.png)

Entität aus Entitätsmenge A steht mit mindestens einer Entität aus Entitätsmenge B in Beziehung. - Entität aus Entitätsmenge B steht mit genau einer Entität aus Entitätsmenge A in Beziehung.

##  8.2 Normalisierung des Entity-Relationship-Diagramms

Datenbanknormalisierung ist ein grundlegender Prozess in der Datenbankgestaltung, der darauf abzielt, Datenstrukturen zu optimieren und Redundanzen zu minimieren, um die Datenintegrität und -Effizienz zu gewährleisten. Durch die Normalisierung werden Daten in logische, miteinander verknüpfte Tabellen aufgeteilt und nach bestimmten Regeln strukturiert, um sicherzustellen, dass Abhängigkeiten und Beziehungen konsistent bleiben. Dieser Prozess ist besonders relevant bei der Gestaltung eines Entity-Relationship-Diagramms (ERD)

### 8.2.1 Normalformen

In der Theorie gibt es fünf Normalformen. Davon sind aber nur drei in der Realität anzutreffen. 

Erste Normalform (1NF): 

Hier werden alle Attribute einer Tabelle in atomare Werte aufgeteilt, das heisst, jede Spalte enthält nur einen Wert. Mehrfachwerte oder Listen sind nicht erlaubt. Dadurch wird sichergestellt, dass die Daten auf einer elementaren Ebene gespeichert sind.

Zweite Normalform (2NF): 

Eine Tabelle ist in der 2NF, wenn sie in der 1NF ist und alle Nichtschlüsselattribute voll funktional vom Primärschlüssel abhängen. Diese Form reduziert Redundanzen, indem sie Attribute in separate Tabellen aufteilt, die nicht vollständig vom Primärschlüssel der Haupttabelle abhängen.

Dritte Normalform (3NF): 

Eine Tabelle erfüllt die 3NF, wenn sie sich in der 2NF befindet und keine transitiven Abhängigkeiten enthält. Das bedeutet, dass Nichtschlüsselattribute nicht voneinander abhängig sein sollten; sie müssen direkt vom Primärschlüssel abhängen.

### 8.2.2 Ablauf Normalisierung

Ein Vorteil ist, dass ein Softwareentwickler im Team die Bedeutung der Normalisierung schon kannte, Dadurch setzte er dies von Anfang an um, damit eine Konsistenz und Flexibilität der Datenstruktur vorhanden ist. Ein erster Schritt, welcher noch im Excel umgesetzt wurde, bestand darin, Adressdaten zu normalisieren, indem diese in eine eigene Tabelle ausgelagert werden. Auf diese Weise können Adressinformationen an zentraler Stelle gepflegt und Dopplungen vermieden werden.

Im Visual Paradigm erfolgte die detaillierte Darstellung der Tabellenbeziehungen, wodurch Abhängigkeiten sichtbar und optimierbar werden. Bei der weiteren Analyse werden ergänzende Aspekte identifiziert und in das ERD-Modell integriert, um die Datenbankstruktur kontinuierlich zu verbessern. Jede Tabelle wurde so modelliert, dass sie die Anforderungen der dritten Normalform grösstenteils erfüllt und keine transitiven Abhängigkeiten bestehen.

Das ERD zeigt die Struktur der Beziehungen zwischen Tabellen wie Hotel, Adresse, Kunde und Buchung. Diese und alle weiteren Tabellen erfüllen die erste Normalform, indem ihre Attribute atomare Werte enthalten. Die zweite Normalform wird eingehalten, indem Attribute, die nicht vom Primärschlüssel der Haupttabelle abhängen, ausgelagert werden. Schliesslich verhindert die dritte Normalform transitive Abhängigkeiten, sodass alle Nichtschlüsselattribute direkt vom Primärschlüssel abhängen.

Die Adressdaten sind beispielsweise in einer separaten Tabelle gespeichert und mit anderen Entitäten wie Kunden Besitzer oder Hotel über Fremdschlüssel verknüpft. Dadurch wird die Wiederverwendbarkeit der Daten ermöglicht und Redundanzen werden vermieden. Ebenso sind Beispielsweise der Pensionstyp, die Features eines Raums, die zusätzlichen Rechnungsartikel und die Zahlungsmethode in einer zusätzlichen Tabelle je nach Beziehung mit einer Zwischentabelle ausgelagert. 

Was hätte noch ausgelagert werden können und weshalb?

In der Tabelle Hotel hätte beispielsweise das Attribut operatingMode noch ausgelagert werden können, da dies von den Werten sehr beschränkt ist und es somit weniger doppelte Einträge gäbe. Wir haben uns jedoch bewusst dagegen entschieden, da wir es lieber etwas übersichtlicher haben wollten und wir schon den Pensionstypen ausgelagert haben. Dies gilt auch für das Attribut type in der Tabelle Room. Desweiteren könnte je nach Erwartung der Datenmenge country in der Tabelle Address ebenso ausgelagert werden mit beispielsweise den Länderkürzel und dem Langtext dazu. Das gleiche gilt beim Attribut currency in der Tabelle Bill. Auch dies könnte ausgelagert werden wenn eine grössere Menge an Daten erwartet wird.

- Diese Tabelle definiert die verschiedenen Arten von Pensionen, die ein Hotel anbieten kann, wie z. B. "Vollpension", "Halbpension" oder "nur Frühstück" dies anhand der Attribute `pensionTypeId` und dem `type`.
- Durch die separate Speicherung von Pensionsarten wird die Datenbank flexibel und erweiterbar; neue Arten können hinzugefügt werden, ohne dass andere Tabellen geändert werden müssen. Dieser Ansatz vermeidet auch Redundanz, da jeder Typ nur einmal definiert wird und von mehreren Hotels referenziert werden kann.



![Room](/bilder/Screenshot_PensionType_table.png)

### Hotel_Pensiontype

- Dient als Brückentabelle zur Darstellung der Many-to-Many-Beziehung zwischen Hotel und Pensiontype.
- Hotels können mehrere Pensionstypen anbieten, und diese Tabelle ermöglicht diese Flexibilität. Sie macht die Datenbankstruktur übersichtlicher und vermeidet Datenduplizierung durch die Verknüpfung von Hotels und Pensionstypen über Fremdschlüssel welche zugleich auch Primärschlüssel sind.

![Room](/bilder/Screenshot_Hotel_PensionType_table.png)

### Rechnung (Bill)

- Diese Tabelle speichert Rechnungsinformationen.
- Die Rechnungstabelle ist für die finanzielle Verfolgung unerlässlich. Denn sie speichert sämtliche relevanten Informationen in den Attributen `billId`, `paymentId`, `bookingId`, `datum`, `währung`, `rechnungsbetrag`, `wurdeBezahlt`.

![Bill](/bilder/Screenshot_Bill_table.png)


### Rechnungsartikel (BillArticles)

- Speichert Rechnungsartikel für die **Rechnung**  mit den Attributen `artikelId`, `beschreibung`, `preis`. 
- Durch die Verknüpfung von **Rechnungsartikel**  über die Tabelle **Rechnung_Rechnungsartikel** zu der Tabelle **Rechnung**  können Rechnungsdetails verknüpft werden. So können beispielsweise verschiedene zusätzliche Angebote im Hotel gebucht und anschliessend mittels einer Rechnung bezhlt werden. 

![BillArticles](/bilder/Screenshot_BillArticles_table.png)

### Rechnung_Rechnungsartikel (Bill_BillArticles)

- Durch die many-to-many Verknüpfung von **Rechnung** und **Rechnungsartikel** entsteht die Tabelle **Rechnung_Rechnungsartikel**. 
-  Verbindet einzelne Artikel oder Dienstleistungen mit einer Rechnung, wobei jeder Artikel eine Beschreibung und einen Betrag hat.
- Diese Struktur ermöglicht eine detaillierte Rechnungsstellung durch Auflistung spezifischer Dienstleistungen oder Produkte, die mit einer Rechnung verbunden sind.

![Bill_BillArticles](/bilder/Screenshot_Bill_BillArticles_table.png)

### Bezahlung (Payment)

- Diese Tabelle legt die für Kunden verfügbaren Zahlungsmethoden fest, mit den Attributen `paymentId` und `Methode` (z. B. PayPal oder Geschenkkarte usw.).

In dieser eignen User Story haben wir zusätzlich das gelernte Wissen genutzt, um ein View zu erstellen und anschliessend auf diesem per Select Statement eine Abfrage getätigt. 

```
import sqlite3
import pandas as pd

# Verbindung zur SQLite-Datenbank herstellen
connection = sqlite3.connect('database/hotel_reservationsystem.db')

# View erstellen
create_view_query = """
CREATE VIEW IF NOT EXISTS Canceled_Bookings_View AS
SELECT 
    Booking.bookingId, 
    Booking.clientId, 
    Booking.bookingDate, 
    Booking.hasBeenCanceled, 
    Client.name || ', ' || Client.surname AS 'Client Name',
    Room.type AS 'Room Type',
    Room.pricePerNight AS 'Price Per Night',
    Hotel.name AS 'Hotel Name',
    Hotel.type AS 'Hotel Type',
    Address.city
FROM 
    Booking
LEFT JOIN 
    Client ON Booking.clientId = Client.clientId
LEFT JOIN 
    Room_Booking ON Booking.bookingId = Room_Booking.bookingId
LEFT JOIN 
    Room ON Room_Booking.roomId = Room.roomId
LEFT JOIN
    Hotel ON Room.hotelId = Hotel.hotelId
INNER JOIN 
    Address ON Hotel.addressId = Address.addressId
WHERE
    Booking.hasBeenCanceled = 1   
GROUP BY 
    Booking.clientId;
"""

cursor = connection.cursor()
cursor.execute(create_view_query)

# Abfrage von dem zuvor erstellten View
view_query = "SELECT * FROM Canceled_Bookings_View;"
result = pd.read_sql_query(view_query, connection)

print(result)

connection.close()
```
